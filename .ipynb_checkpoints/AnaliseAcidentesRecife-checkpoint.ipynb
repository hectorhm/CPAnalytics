```python
#imports
import pandas as pd
import numpy as np
```

```python
#carregando dados sobre semaforos da cidade do Recife
semaforos = pd.read_csv(
    "http://dados.recife.pe.gov.br/dataset/ab87d3e2-ea78-4f27-95fd-d7ef6f3b9ee5/resource/ab6343e9-c3f2-4d62-9554-5778f9f33738/download/semaforos.csv",
    sep=';',header=None, names=["semaforo", "localizacao1", "localizacao2", "funcionamento", "utilizacao", "sinalsonoro", "sinalizadorciclista", "Latitude", "Longitude"])
semaforosDG = semaforos.drop(['funcionamento'], axis=1)
semaforosDG
```

```python
#Carrega dados sobre acidentes com vítimas no ano de 2016 e 2017
acidentesComVitimas2k15 = pd.read_csv(
    "http://dados.recife.pe.gov.br/dataset/160c5176-f56e-4ca9-b746-a6a6cec3dd8b/resource/eb33c6b5-5708-4d95-a413-0e9de855b198/download/acidentes-2015.csv",
    sep=';',header=None, names=["data","tipo", "detalhes", "longitude", "latitude"])
acidentesComVitimas2k16 = pd.read_csv(
    "http://dados.recife.pe.gov.br/dataset/61545107-2e7d-4506-b9fa-f5264a7f6ec9/resource/d153f88e-3c25-422b-8b94-ef8d660bf7bf/download/acidentes-2016.csv",
    sep=';',header=None, names=["data","tipo", "detalhes", "longitude", "latitude"])
acidentesComVitimas = pd.concat([acidentesComVitimas2k15, acidentesComVitimas2k16])
acidentesComVitimas
```
